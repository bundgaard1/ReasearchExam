```python
import pandas as pd

demographic_filepath = "merged_data.xlsx - Demographics.csv"
usability_filepath = "merged_data.xlsx - Usability_Summary.csv"

demo_data = pd.read_csv(demographic_filepath)
usability_data = pd.read_csv(usability_filepath)

headers = demo_data.keys()
demo_data
```

# Group demographic Profile

```python
import matplotlib.pyplot as plt

age = demo_data["Q1) What is your age?"]
gender = demo_data["Q2) What is your gender?"]
education = demo_data[headers[3]]
experience = demo_data["Q4) Have you ever used GenerativeAI (GenAI)? (Yes/No)?"]
ai_usage = demo_data["Q5) If you answered \'Yes\' to Q4, how often do you use GenAI? (e.g. once a week)"]
nationality = demo_data["Q7) Which country you feel most connected to? This may not be the country where you were born"]

```

```python
# Count occurrences of each age
age_counts = age.value_counts().sort_index()

plt.bar(age_counts.index, age_counts.values)
plt.xlabel("Age")
plt.ylabel("Count")
plt.title("Age Distribution")
plt.show()
```

```python
BEng_electronics = ["Applied BSc in Electronics", "B.Eng. in Electrical Engineering", "BEng Electrical Engineering (Electronics)", "BEng in Electrical Engineering" ]
EEng_healthteck= ["B.Sc. in Engineering in Health Technology", ]
BEng_software= ["BEng in Software Technology Engineering", "BSc Software Eng."]
BSc_computer = ["BS Electrical Computer Engineering", "BSc in Computer Engineering", "BSc in Computer Science", "BSc in Computer Science Engineering", "BSc in Computer Scinece and Engineering"]
BSc_electro = ["BSC in Electrical Engineering", "BSc Eletrical Engineering", "BSc in Electrical Engineering", "BSc in Electronical Engineering", "BSc in Electronincs Engineering", "BSc in Elektroteknologi"]
BSc_AI = ["BSC in AI"]
BSc_audiovisual = ["BSc Audiovisual Systems Engineering",]

# merge similar education levels
education_merged = education.replace(BEng_electronics, "BEng in Electronics")
education_merged = education_merged.replace(EEng_healthteck, "BSc in Health Technology Engineering")
education_merged = education_merged.replace(BEng_software, "BEng in Software Engineering")
education_merged = education_merged.replace(BSc_computer, "BSc in Computer Engineering")
education_merged = education_merged.replace(BSc_electro, "BSc in Electrical Engineering")
education_merged = education_merged.replace(BSc_AI, "BSc in AI")   
education_merged = education_merged.replace(BSc_audiovisual, "BSc in Audiovisual Systems Engineering")

education_counts = education_merged.value_counts().sort_index()

plt.barh(education_counts.index, education_counts.values,)
plt.xlabel("Education Level")

```
